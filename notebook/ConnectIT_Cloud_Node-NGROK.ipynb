```python
# @title 2. Configure Ngrok Tunnel
# @markdown Enter your Ngrok Authtoken below. Get it from [dashboard.ngrok.com](https://dashboard.ngrok.com).

import os
from pyngrok import ngrok, conf

NGROK_TOKEN = "2o09Xt46sImUT417zQicmBAL1Rx_5ZHWep4o6rGtH6oEa9V9T" # @param {type:"string"}
PORT = 4003 # @param {type:"integer"}

if NGROK_TOKEN:
    conf.get_default().auth_token = NGROK_TOKEN
    # Kill existing tunnels
    ngrok.kill()
    # Open a HTTP tunnel on the port
    public_url = ngrok.connect(PORT, proto="http").public_url
    # Convert http/https to ws/wss for ConnectIT
    public_addr = public_url.replace("http://", "ws://").replace("https://", "wss://")
    print(f"‚úÖ Tunnel Active! Public Address: {public_addr}")
    
    # Set ENV variables for ConnectIT to see
    os.environ["CONNECTIT_PUBLIC_ADDR"] = public_addr
    os.environ["CONNECTIT_PORT"] = str(PORT)
else:
    print("‚ùå Please provide an Ngrok Token to accessible outside Colab.")
```
